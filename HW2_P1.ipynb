```python
def generate_image(img):
    rows, cols, ch = img.shape

    #Random Scale
    scale_factor = np.random.uniform(0.5, 1.5)

    # Random rotation
    random_rot = np.random.uniform(-30, 30)
    Rot_M = cv2.getRotationMatrix2D((cols/2, rows/2), random_rot, scale_factor)

    # Random translation
    tr_x = np.random.uniform(-20, 20)
    tr_y = np.random.uniform(-20, 20)
    Trans_M = np.float32([[1, 0, tr_x], [0, 1, tr_y]])

    # Random shear
    pts1 = np.float32([[5, 5], [20, 5], [5, 20]])
    pt1 = 5 + np.random.uniform(-5, 5)
    pt2 = 20 + np.random.uniform(-5, 5)
    pts2 = np.float32([[pt1, 5], [pt2, pt1], [5, pt2]])
    Shear_M = cv2.getAffineTransform(pts1, pts2)



    # Apply transformations
    img = cv2.warpAffine(img, Rot_M, (cols, rows))
    img = cv2.warpAffine(img, Trans_M, (cols, rows))
    img = cv2.warpAffine(img, Shear_M, (cols, rows))

    return img
```

```python
gs1 = gridspec.GridSpec(10, 10)
gs1.update(wspace=0.01, hspace=0.02)  # set the spacing between axes.
plt.figure(figsize=(12, 12))

for i in range(100):
    transformed_image = generate_image(img.copy())  # Apply transformations

    ax = plt.subplot(gs1[i])
    ax.imshow(cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB))  # Assuming OpenCV's BGR format
    ax.axis('off')

plt.show()
```

```python
def zca_grey(patch):
    
    num_patch, width, height = patch.shape
    print(patch.shape)
    print(num_patch)
    x = patch.reshape(num_patch, width * height).T 
    print('here')
    print(x.shape)

    avg = np.mean(x, axis = 1, keepdims=True)
    x= x-avg
    cov = np.dot(x,x.T)/x.shape[1]

    U, S, V = np.linalg.svd(cov)

    xRot = np.dot(U.T,x)

    k =1
    xTilde = np.dot(U[:,:k].T,x)

    epsilon = 0.1
    xZCAwhite = np.dot(U, np.dot(np.diag(1.0 / np.sqrt(S + epsilon)), np.dot(U.T, x)))

    xZCAwhite_patches = xZCAwhite.T.reshape(num_patch, width, height)

    #print(xZCAwhite_patches.shape)

    return(xZCAwhite_patches)
    
    # Plotting the covariance matrix
    # plt.figure(figsize=(10, 10))
    # vmax = np.abs(sigma_rot).max()  # Adjust color scale based on the maximum absolute value
    # plt.imshow(sigma_rot, cmap='hot', interpolation='nearest', vmax=vmax)
    # plt.colorbar()
    # plt.title('Covariance Matrix of Rotated Data')
    # plt.show()

    

def zca_color(patch):
    
    #print(patch.shape)
    num_patches, width, height, num_channels = patch.shape
    # Initialize an array to store the ZCA whitened patches
    xZCAwhite_patches_all_channels = np.empty((num_patches, width, height, num_channels)) 
    
    for channel in range(num_channels):
        current_channel = patch[:,:,:,channel]

        x = current_channel.reshape(num_patches, width * height).T

        #print(x.shape)
        avg = np.mean(x, axis=1, keepdims=True)
        x - x- avg

        cov = np.dot(x, x.T) / x.shape[1]

        # SVD
        U, S, V = np.linalg.svd(cov)

        epsilon = 0.1
        xZCAwhite = np.dot(U, np.dot(np.diag(1.0 / np.sqrt(S + epsilon)), np.dot(U.T, x)))

        xZCAwhite_min = xZCAwhite.min(axis=0, keepdims=True)
        xZCAwhite_max = xZCAwhite.max(axis=0, keepdims=True)
        xZCAwhite = (xZCAwhite - xZCAwhite_min) / (xZCAwhite_max - xZCAwhite_min)
        

        xZCAwhite_patches = xZCAwhite.T.reshape(num_patches, width, height)
        xZCAwhite_patches_all_channels[:,:,:,channel] = xZCAwhite_patches

    return xZCAwhite_patches_all_channels
```

```python
ZCAwhite = zca_grey(np_grey_patches)
ZCAcolor = zca_color(np_color_patches)
plt.close()
for i in range(0,3):
    plt.imshow(np_color_patches[i])
    plt.show()
    plt.close()

    plt.imshow(ZCAcolor[i])
    plt.savefig("./figures/ZCAcolor" + str(i) + ".png", bbox_inches='tight', pad_inches=0)

    plt.show()
    plt.close()

    
    
    plt.imshow(np_grey_patches[i], cmap='gray')
    plt.show()
    plt.close()

    plt.imshow(ZCAwhite[i], cmap='gray')
    plt.savefig("./figures/ZCAgrey" + str(i) + ".png", bbox_inches='tight', pad_inches=0)
    plt.show()
    plt.close()

```
